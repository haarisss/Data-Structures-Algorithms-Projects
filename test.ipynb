```python
from turtle import Turtle

class runner(Turtle):
    def __init__(self,x,y):
        super().__init__()
        self.hideturtle()
        self.color('red')
        self.penup()
        self.speed('slow')
        self.setheading(90)
        self.goto(x,y)
        self.showturtle()
        # self.pendown()

    # def go_left(self,length):
    #     #go left
    #     self.setheading(180)  #Face West
    #     self.forward(length)

    # def go_right(self,length):
    #     #go right
    #     self.setheading(0) #Face East
    #     self.forward(length)

    # def go_up(self,length):
    #     #go up
    #     self.setheading(90)  #Face North
    #     self.forward(length)

    # def go_down(self,length):
    #     #go down
    #     self.setheading(270)    #Face South
    #     self.forward(length)

import networkx as nx

class dijkstra(runner):
    def __init__(self,start_x,start_y,graph):
        self.start_x = start_x
        self.start_y = start_y
        self.graph = graph

    def shortest_path(self,target):
        g = self.graph
        source = (self.start_x,self.start_y)
        path = nx.shortest_path(g,source,target)
        length = nx.shortest_path_length(g,source,target)

        return path,length

    def move(self,target):
        x,y = target
        target = (x,y)
        path, length = self.shortest_path(target)
        print(f'Shortest Path: {path}\nPath Length: {length}')
        for step in path:
            print(step[0],step[1])
```
